```python
#Fledgling data from MongoDB
data_list = []
for data in collection.find({},{"_id": 0}):
  data_list.append(data)

df = pd.DataFrame(data_list)
```

```python
df['last_scraped'] = pd.to_datetime(df['last_scraped'])
df['calendar_last_scraped'] = pd.to_datetime(df['calendar_last_scraped'])
df['first_review'] = pd.to_datetime(df['first_review'])
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df['minimum_nights'] = pd.to_numeric(df['minimum_nights'])
df['maximum_nights'] = pd.to_numeric(df['maximum_nights'])
df['bathrooms'] = pd.to_numeric(df['bathrooms'], errors='coerce')
```

```python
df['amenities'] = df['amenities'].apply(lambda x: ', '.join(x))
```

```python
for i in df.columns:
  print(i, " : ", df[i].duplicated().sum())
```
